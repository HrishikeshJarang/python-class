```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv('E:\\HRISHIKESH\\AWS\\Data Analysis\\Documents\\Excel\\Data Visualization\\bar_chart_data.csv')
df_used_cars
```

```python
plt.figure(figsize = (6,6))
plt.bar(x = df_used_cars['Brand'],  #x-axis = Brand
       height = df_used_cars['Cars Listings'],
       color = 'pink')
plt.ylabel('Number of Listings', fontsize = 10) #y-axis = Number of Listings
plt.xticks(rotation = 45, fontsize = 10)  #Orientation on x-axis
plt.yticks(fontsize = 10)
plt.title('--Cars Listings by Brand--', 
          fontsize = 15, 
         fontweight = 'bold')
plt.show()
plt.savefig('Used Cars Bar.png')
```
